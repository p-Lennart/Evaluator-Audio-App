```python
import pandas as pd
```

```python
aotgs = pd.read_csv("../public/air_on_the_g_string/baseline/aotgs_tempo.csv")
```

```python
aotgs.head()
```

```python
out_path = "data/test_suite/07_aotgs_ground_truth.txt"

def midi_to_freq(m):
    return 440.0 * 2 ** ((m - 69) / 12)

step = 0.005
t_min = aotgs["ref_ts"].min()
t_max = aotgs["ref_ts"].max()
n_steps = int(round((t_max - t_min) / step))

written = 0
with open(out_path, "w") as f:
    for i in range(n_steps + 1):
        t = t_min + i * step
        idx = aotgs["ref_ts"].searchsorted(t, side="right") - 1
        if idx < 0:
            continue
        m = int(aotgs.iloc[idx]["midi"])
        f.write(f"{t:.6f}\t{midi_to_freq(m):.2f}\n")
        written += 1

print(f"Wrote {written} lines to {out_path}")
```
